```python
#imports
import pandas as pd
from sklearn.preprocessing import StandardScaler, normalize, scale
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy import stats
```

```python
# Download the data
link = 'https://think.cs.vt.edu/corgis/datasets/csv/cars/cars.csv'
data = pd.read_csv(link)
data.head()
```

```python
# Download the data
list(data.columns.values)
```

```python
# Feature engineering
df = data[(data[:]['Identification.Make'].str.contains('Honda')) & data[:]['Fuel Information.Fuel Type'].str.contains('Gasoline')].select_dtypes('number')
df
```

```python
#mpg1 = df[df['Fuel Information.Highway mpg'] < df['Fuel Information.Highway mpg'].quantile(0.90) < df['Fuel Information.Highway mpg']]
newdf = df.loc[:, ['Fuel Information.Highway mpg', 'Engine Information.Engine Statistics.Horsepower']]

mpgoutliers = newdf[newdf['Fuel Information.Highway mpg'] < newdf['Fuel Information.Highway mpg'].quantile(0.90)]
mpg199 = mpgoutliers[mpgoutliers['Engine Information.Engine Statistics.Horsepower'] < mpgoutliers['Engine Information.Engine Statistics.Horsepower'].quantile(0.90)]


#mpgoutliers1 = mpg199[mpg199['Fuel Information.Highway mpg'] < mpg199['Fuel Information.Highway mpg'].quantile(0.90)]
#mpg1991 = mpgoutliers1[mpgoutliers1['Engine Information.Engine Statistics.Horsepower'] < mpgoutliers1['Engine Information.Engine Statistics.Horsepower'].quantile(0.90)]
#newdf
mpg199
#newdf['Fuel Information.Highway mpg'].quantile(0.90)
```

```python
# scaler = StandardScaler()

# mpg200 = pd.DataFrame(scaler.fit_transform(mpg199), columns = mpg199.columns)
# mpg200_col = mpg200.columns
mpg201 = normalize(mpg199)
mpg202 = pd.DataFrame(mpg201, columns=mpg199.columns)
mpg202
```

```python
# lets use hgrad and unem as predictors and roll as predictant
# enroll_data = mpg202[['Engine Information.Engine Statistics.Horsepower','Fuel Information.Highway mpg']].values
# target = enroll['roll'].values
# enroll_data_names = ['unem','hgrad']
# x,y = scale(enroll_data),target # scale is removing the mean value of each feature, then scale it by dividing features by their standard deviation
# model = LinearRegression(normalize=True)
# model.fit(x,y)
# print(model.score(x,y)) # score is the degree of linear correlation between data and target
x = mpg202['Engine Information.Engine Statistics.Horsepower']
y = mpg202['Fuel Information.Highway mpg']
model = LinearRegression()
xs = np.array(x).reshape(-1, 1)
ys = np.array(y).reshape(-1, 1)
model.fit(ys,x)
print(model.score(ys,xs))
#mpg202.plot.scatter(x = 'Engine Information.Engine Statistics.Horsepower', y = 'Fuel Information.Highway mpg')
plt.scatter(x,y,color="g")
plt.plot(x,model.predict(xs) )
```
